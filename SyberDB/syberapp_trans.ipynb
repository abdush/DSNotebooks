```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np

#create engine to DB. echo flag is used to enable/disable logging and show generated sql.
engine = create_engine('mysql+pymysql://admin:admin@localhost:3307/analysis', echo=False)
trans = pd.read_sql_query('SELECT * FROM app_trans_log', engine, index_col = 'id')
```

```python
#how many rows and cols
print (trans.shape)
# row labels (indexes shown on the left of the data shown below)
print (trans.index)
#pring the first 5 rows. same as users.iloc[:5,:]
trans.head()
```

```python
# show col data types
trans.dtypes
```

```python
# check unique values of some cols
print("Status unique values:")
print(trans['status'].unique())
print()

# get the values count for status col
print("Status value counts (including NA values):")
print(trans['status'].value_counts(dropna=False))
print()

# get the values count for tran_type col
print("Trans Types value counts (including NA values):")
print(trans['tran_type'].value_counts(dropna=False))
print()

# get the values count for clientId col
print("Client IDs value counts (including NA values):")
print(trans['client_id'].value_counts(dropna=False))
```
